```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans

# Chargement des données Iris
iris = load_iris()
X = iris.data
y = iris.target

# Entraînement du modèle K-means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Obtention des centroïdes et des étiquettes de cluster prédites
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Affichage des résultats
plt.figure(figsize=(10, 6))

# Affichage des points de données
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5, label='Data Points')

# Affichage des centroïdes
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', c='red', s=200, label='Centroids')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-means Clustering on Iris Dataset')
plt.legend()
plt.show()
```
